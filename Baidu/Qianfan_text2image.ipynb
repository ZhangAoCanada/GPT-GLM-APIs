以下是一个简单的文生图的案例，使用的千帆SDK版本为：
```
qianfan >= 0.1.2
```

````python
import os
from dotenv import load_dotenv

load_dotenv()
````

````python
import qianfan
from PIL import Image 
import io

t2i = qianfan.Text2Image()
resp = t2i.do(prompt="An super high-resolution and realistic image captured on the road: trucks running on highway when rainy.", with_decode="base64", steps=50)
img_data = resp["body"]["data"][0]["image"]

img = Image.open(io.BytesIO(img_data))
display(img)
````
